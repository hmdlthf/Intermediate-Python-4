```python
import  matplotlib as mpl
from matplotlib import pyplot as plt
```

```python
slices = [60, 40]
```

```python
labels = ["sixty", "forty"]
plt.pie(slices, labels=labels)

plt.title("My Pie Chart")
plt.tight_layout()
plt.show
```

```python
slices = [60, 40, 30, 10]
labels = ["sixty", "forty", "thirty", "ten"]
colors = ["magenta", "cyan", "yellow", "red"]
```

```python
plt.pie(slices, labels=labels, colors=colors, wedgeprops={'edgecolor': "black"})

plt.title("My Pie Chart")
plt.tight_layout()
plt.show
```

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
```

```python
plt.pie(slices, labels=labels)

plt.title("My Pie Chart")
plt.tight_layout()
plt.show
```

```python
explode = [0.1,0.1,0.5,0.1,0.1]
plt.pie(slices, labels=labels, explode=explode)

plt.title("My Pie Chart")
plt.tight_layout()
plt.show
```

```python
explode = [0,0,0.1,0,0]
plt.pie(slices, labels=labels, explode=explode,autopct= '%1.1f%%')

plt.title("My Pie Chart")
plt.tight_layout()
plt.show
```

```python
explode = [0,0,0.1,0,0]
plt.pie(slices, labels=labels, explode=explode,autopct= '%1.1f%%', shadow= True)

plt.title("My Pie Chart")
plt.tight_layout()
plt.show
```

```python
explode = [0,0,0.1,0,0]
plt.pie(slices, labels=labels, explode=explode,autopct= '%1.1f%%', shadow= True, startangle=270)

plt.title("My Pie Chart")
plt.tight_layout()
plt.show
```
